```python
# Import pandas
import pandas as pd
```

```python
# Read the orders.csv file
data_df = pd.read_csv("orders.csv")

data_df.head(10)
```

```python
# Search for null values
data_df["Ship Mode"].unique()
```

```python
# Handle null values
data_df = pd.read_csv("orders.csv", na_values=["Not Available", "unknown"])
data_df["Ship Mode"].unique()
```

```python
# Rename columns, make them lower case and replace space with underscore
data_df.columns = data_df.columns.str.lower()
data_df.columns = data_df.columns.str.replace(" ", "_")
data_df.head(10)
```

```python
# Derive new columns: discount_price, sale_price, and profit
data_df["discuount"] = data_df["list_price"] * data_df["discount_percent"]* 0.01
data_df["sale_price"] = data_df["list_price"] - data_df["discuount"]
data_df["profit"] = data_df["sale_price"] - data_df["cost_price"]
data_df.head(10)
```

```python
# Review dtypes
data_df.dtypes
```

```python
# Convert order_date column to date type
data_df["order_date"] = pd.to_datetime(data_df["order_date"],format = "%Y-%m-%d")
data_df.head(5)
```

```python
# Review dtypes
data_df.dtypes
```

```python
# Drop cost_price, list_price, and discount_percent columns
data_df.drop(columns=["cost_price", "list_price", "discount_percent"],inplace=True)
data_df.head(10)
```

```python
# Load the data into SQL server using replace option
import sqlalchemy as sal
engine = sal.create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/SQL_Python_Project")
conn = engine.connect()
```

```python
data_df = data_df.rename(columns={"discuount": "discount"})
```

```python
# Load the data into sql server using append option
data_df.to_sql("df_orders", con=conn, index=False, if_exists= "append")
```
